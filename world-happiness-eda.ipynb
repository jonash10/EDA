```python
import numpy as np 
import pandas as pd 
import os
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/kaggle/input/world-happiness/2019.csv')
```

```python
df.head()
```

# Which Factor correlates closest to the Happiness score

### Social Support

```python
# Scatter plot 
plt.figure(figsize=(8,6))
plt.scatter(df['Social support'], df['Score'], color='blue')

plt.title('Score vs Social support')
plt.xlabel('Social support')
plt.ylabel('Score')
plt.grid(True)
plt.show()
```

If there is a perfect correlation between two factors, then the scatter graph will present a linear relationship between the two variables.
Within this dataset there is not any individual factor that correlates exactly with the happiness score.
Factors that come close to having a linear relationship are; GDP per capita, social support and healthy life expectancy
