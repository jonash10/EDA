# Video Game sales EDA

```python
df = pd.read_csv('/kaggle/input/videogamesales/vgsales.csv')
```

```python
df.head()
```

```python
# Sort by 'Global_Sales'
df_sales = df.sort_values(by='Global_Sales', ascending=False)

# display top ten
df_sales.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#find the topten
top_10_games = df_sales.head(10)

# the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Global_Sales', y='Name', data=top_10_games, palette='viridis')

# Description
plt.title('Top 10 Games by Global Sales', fontsize=14)
plt.xlabel('Global Sales (in millions)', fontsize=12)
plt.ylabel('Game Name', fontsize=12)

# plotting
plt.show()
```
