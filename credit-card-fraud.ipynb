# What are the characteristics of the dataset


```python
df_training = train.drop('Time', axis=1)
df_testing = test.drop('Time', axis=1)
```

```python

l1 = list(df_training['Class'].value_counts())
pie_values = [l1[0] / sum(l1) * 100, l1[1] / sum(l1) * 100]

# Pie chart
plt.figure(figsize=(8, 7))
plt.pie(pie_values, labels=['Valid', 'Fraud'], 
        autopct='%1.3f%%', startangle=90, 
        explode=(0.05, 0.05), colors=["#2196F3", "#FF9800"],
        wedgeprops={'edgecolor': 'black', 'linewidth': 1, 'antialiased': True})
plt.title('Valid transactions and Fraudulent transactions')
plt.show()
```

This Dataset appears to contain 99.8% genuine transactions and only 0.2% fraudulent transactions.
