# Movie analysis

```python
df = pd.read_csv("/kaggle/input/tmdb-movie-metadata/tmdb_5000_movies.csv")
```

```python
df.head()
```

```python
# unique values from genres
unique_values = df['genres'].unique()

print(unique_values)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['popularity'], df['budget'], color='blue')

# Filter to movies under 200 popularity
plt.xlim(0, 200)

plt.title('budget vs popularity')
plt.xlabel('popularity')
plt.ylabel('budget (1 = 100 million)')
plt.grid(True)
plt.show()
```

There is one big cluster at the lower end of the x and y axis, this is the cluster that contains low budget low popularity films.

One other interesting thing to note is that the films in the 125-150 popularity rating zone remains fairly evenly spread, from low budget to high budget.

This graph reveals that while it is possible to spend a lot and make a universally disliked film, if you are spending above 100 million it is likely your film will be well recieved.
