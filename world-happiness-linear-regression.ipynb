```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# Download and prepare the data
df = pd.read_csv('/kaggle/input/world-happiness/2019.csv')

X = df[["GDP per capita"]].values
y = df[["Score"]].values
```

```python
# Data visualisation 
df.plot(kind='scatter', grid=True,
x="GDP per capita", y="Score")
plt.axis([0, 2, 4, 9])
plt.show()
```

```python
# Model selection
model = LinearRegression()
```

```python
# Model Training
model.fit(X, y)
```

```python
# Predicting Costa Rica's score
X_new = [[1.34]] 
print(model.predict(X_new)) 
```
